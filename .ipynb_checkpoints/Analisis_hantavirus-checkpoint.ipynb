```python
import pandas as pd
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh.palettes import Category20
output_notebook()
import pygwalker as pyg
warnings.filterwarnings("ignore")
```

```python
# Carga del CSV
```

```python
df = pd.read_csv("hantavirus.csv")
print(f"CSV cargado: {len(df)} registros, {len(df.columns)} columnas")
```

```python
# Creación de la base de datos SQLite
```

```python
conn = sqlite3.connect("hantavirus.db")
```

```python
# Tabla principal
```

```python
df.to_sql("casos", conn, if_exists="replace", index=False)
print("Tabla 'casos' creada")
```

```python
# Tabla resumen por país
```

```python
resumen = df.groupby("country").agg(
    total_casos=("confirmed_cases", "sum"),
    total_muertes=("deaths", "sum"),
    total_recuperados=("recovered", "sum"),
    promedio_letalidad=("case_fatality_rate", "mean"),
    años_registrados=("year", "nunique")
).reset_index()
```

```python
resumen.to_sql("resumen_pais", conn, if_exists="replace", index=False)
print("Tabla 'resumen_pais' creada")
```

```python
# Guardar la base de datos
```

```python
conn.close()
print("\n🗄️  Base de datos guardada: hantavirus.db")
```

```python
# proceso de carga y merge de datos
```

```python
df_csv = pd.read_csv("hantavirus.csv")
```

```python
conn = sqlite3.connect("hantavirus.db")
df_db  = pd.read_sql("SELECT * FROM casos", conn)
df_res = pd.read_sql("SELECT * FROM resumen_pais", conn)
conn.close()
```

```python
df_final = df_csv.merge(df_res, on="country", how="left")

print(f" df_csv  → {df_csv.shape[0]} filas x {df_csv.shape[1]} columnas")
print(f" df_db   → {df_db.shape[0]} filas x {df_db.shape[1]} columnas")
print(f" df_res  → {df_res.shape[0]} filas x {df_res.shape[1]} columnas")
print(f" df_final → {df_final.shape[0]} filas x {df_final.shape[1]} columnas")

print("\n Vista previa del dataset final:")
df_final.head(3)
```

```python
# 1ra visualización MATPLOTLIB
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
datos = df_final.groupby("country")["confirmed_cases"].sum().sort_values(ascending=True)
bars = ax.barh(datos.index, datos.values, color="steelblue", edgecolor="white")
for bar in bars:
    width = bar.get_width()
    ax.text(width + 500, bar.get_y() + bar.get_height()/2,
            f"{width:,.0f}", va="center", fontsize=8)
ax.set_title("Casos Confirmados de Hantavirus por País (1950–2025)",
             fontsize=14, fontweight="bold", pad=15)
ax.set_xlabel("Total de Casos Confirmados", fontsize=11)
ax.set_ylabel("País", fontsize=11)
ax.xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f}"))
ax.grid(axis="x", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 2da visualización
```

```python
fig, ax = plt.subplots(figsize=(14, 5))
por_año = df_final.groupby("year")["confirmed_cases"].sum()
ax.plot(por_año.index, por_año.values, color="steelblue",
        linewidth=2, marker="o", markersize=3)
ax.fill_between(por_año.index, por_año.values, alpha=0.15, color="steelblue")
ax.fill_between(por_año.index, por_año.values, alpha=0.15, color="steelblue")
ax.set_title("Evolución Anual de Casos Confirmados de Hantavirus (1950–2025)",
             fontsize=13, fontweight="bold", pad=15)
ax.set_xlabel("Año", fontsize=11)
ax.set_ylabel("Casos Confirmados", fontsize=11)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f}"))
ax.grid(linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Visualizaión 3
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
por_region = df_final.groupby("who_region")[["deaths", "recovered"]].sum()
x = range(len(por_region.index))
ancho = 0.35
barras1 = ax.bar([i - ancho/2 for i in x], por_region["deaths"],
                 width=ancho, label="Muertes", color="crimson", edgecolor="white")
barras2 = ax.bar([i + ancho/2 for i in x], por_region["recovered"],
                 width=ancho, label="Recuperados", color="seagreen", edgecolor="white")
for bar in barras1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
            f"{bar.get_height():,.0f}", ha="center", fontsize=9)
for bar in barras2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 500,
            f"{bar.get_height():,.0f}", ha="center", fontsize=9)
ax.set_title("Muertes vs Recuperados por Región OMS",
             fontsize=13, fontweight="bold", pad=15)
ax.set_xlabel("Región OMS", fontsize=11)
ax.set_ylabel("Total de Personas", fontsize=11)
ax.set_xticks(list(x))
ax.set_xticklabels(por_region.index, fontsize=11)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{x:,.0f}"))
ax.legend(fontsize=10)
ax.grid(axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
# 4ta visualización
```

```python
df_final["decada"] = (df_final["year"] // 10) * 10
pivot = df_final.groupby(["country", "decada"])["confirmed_cases"].sum().unstack(fill_value=0)
fig, ax = plt.subplots(figsize=(14, 9))
im = ax.imshow(pivot.values, aspect="auto", cmap="YlOrRd")
ax.set_xticks(range(len(pivot.columns)))
ax.set_xticklabels([str(d) for d in pivot.columns], fontsize=10)
ax.set_yticks(range(len(pivot.index)))
ax.set_yticklabels(pivot.index, fontsize=9)
for i in range(len(pivot.index)):
    for j in range(len(pivot.columns)):
        val = pivot.values[i, j]
        if val > 0:
            ax.text(j, i, f"{val:,.0f}", ha="center", va="center",
                    fontsize=7, color="black" if val < pivot.values.max()*0.6 else "white")
plt.colorbar(im, ax=ax, label="Casos Confirmados")
ax.set_title("Casos Confirmados de Hantavirus por País y Década",
             fontsize=13, fontweight="bold", pad=15)
ax.set_xlabel("Década", fontsize=11)
ax.set_ylabel("País", fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# 5ta Visualización BOKEH
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool, ColumnDataSource
from bokeh.palettes import Category20
output_notebook()
top10 = (df_final.groupby("country")["confirmed_cases"]
         .sum()
         .sort_values(ascending=False)
         .head(10)
         .reset_index())
top10.columns = ["country", "confirmed_cases"]
top10 = top10.sort_values("confirmed_cases", ascending=True)
source = ColumnDataSource(top10)

p = figure(y_range=top10["country"].tolist(),
           height=400, width=750,
           title="Top 10 Países con Más Casos de Hantavirus",
           toolbar_location="above")

p.hbar(y="country", right="confirmed_cases", height=0.6,
       source=source, color="steelblue", alpha=0.8)

hover = HoverTool(tooltips=[
    ("País", "@country"),
    ("Casos", "@confirmed_cases{,}")
])
p.add_tools(hover)
p.xaxis.formatter.use_scientific = False
p.xgrid.grid_line_color = "lightgray"
p.xgrid.grid_line_dash = "dashed"
p.title.text_font_size = "13px"
show(p)
```

```python
# 6ta visualización
```

```python
from bokeh.models import Legend
por_sindrome = (df_final.groupby(["year", "syndrome"])["confirmed_cases"]
                .sum()
                .reset_index())
hps  = por_sindrome[por_sindrome["syndrome"] == "HPS"]
hfrs = por_sindrome[por_sindrome["syndrome"] == "HFRS"]
source_hps  = ColumnDataSource(hps)
source_hfrs = ColumnDataSource(hfrs)
p2 = figure(height=400, width=800,
            title="Evolución de Casos por Síndrome (HPS vs HFRS)",
            toolbar_location="above")
l1 = p2.line("year", "confirmed_cases", source=source_hps,
             line_color="steelblue", line_width=2, legend_label="HPS (Pulmonar)")
l2 = p2.line("year", "confirmed_cases", source=source_hfrs,
             line_color="darkorange", line_width=2, legend_label="HFRS (Renal)")
hover2 = HoverTool(tooltips=[
    ("Año",      "@year"),
    ("Síndrome", "@syndrome"),
    ("Casos",    "@confirmed_cases{,}")
])
p2.add_tools(hover2)
p2.legend.location = "top_left"
p2.legend.click_policy = "hide"
p2.xgrid.grid_line_dash = "dashed"
p2.ygrid.grid_line_dash = "dashed"
p2.title.text_font_size = "13px"
p2.yaxis.formatter.use_scientific = False
show(p2)
```
